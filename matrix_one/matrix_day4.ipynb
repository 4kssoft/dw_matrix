```
df['prices_condition_cat'] = df.prices_condition.str.lower().factorize()[0]
```

```
from  sklearn.metrics import SCORERS
SCORERS.keys()
```

```
feats = ['brand_cat']
X = df[feats].values
y = df['prices_amountmin'].values

model = DecisionTreeRegressor(max_depth=5)

scores = cross_val_score(model, X, y, scoring='neg_mean_absolute_error')
np.mean(scores), np.std(scores)
```

```
def run_model(feats):
  X = df[feats].values
  y = df['prices_amountmin'].values

  model = DecisionTreeRegressor(max_depth=5)

  scores = cross_val_score(model, X, y, scoring='neg_mean_absolute_error')
  return np.mean(scores), np.std(scores)
```

```
run_model(['brand_cat'])
```

```
run_model(['manufacturer_cat'])
```

```
run_model(['prices_condition_cat'])
```

```
run_model(['brand_cat','prices_condition_cat'])
```

```
run_model(['brand_cat','manufacturer_cat','prices_condition_cat'])
```
